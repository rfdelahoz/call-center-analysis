### 2.4 Diagrama de dispersi칩n
El diagrama de dispersi칩n nos permite ver relaciones y detectar algunos **patrones** entre las dos variables. En este estudio tomaremos como
* variable independiente $X =$ **N칰mero de llamadas entrantes** 
* Variable dependiente $Y=$ **Tiempo de espera promedio (Segundos)**  

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Asumiendo que tus datos est치n en new_df
x1 = new_df["Llamadas entrantes"]
y1 = new_df["Tiempo de espera promedio (Segundos)"]

# Crear el gr치fico de dispersi칩n con regresi칩n lineal
plt.figure(figsize=(10, 6))
sns.regplot(x=x1, y=y1, scatter_kws={'alpha': 0.5, 'color': 'blue','edgecolor':'black'}, line_kws={'color': 'red'})

# Personalizar el gr치fico
plt.title("Relaci칩n entre Llamadas entrantes y Tiempo de espera promedio", fontsize=14)
plt.xlabel("Llamadas entrantes", fontsize=12)
plt.ylabel("Tiempo de espera promedio (Segundos)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Mostrar el gr치fico
plt.show()
````

Del gr치fico anterior, podemos observar alg칰n patr칩n o tendencia lineal positiva. Es decir, una tendencia lineal creciente.  Sin embargo, necesitamos cuantificar y medir este patr칩n, lo cu치l se har치 con una medida muy conocida, llamada:  **coeficiente de correlaci칩n lineal**.

### 2.5 Coeficiente de correlaci칩n
Necesitamos medir la **fuerza**  de relaci칩n que existe entre nuestras variables de estudio. Por lo tanto, una manera de medir dicha fuerza, es utilizando el **coeficiente de correlaci칩n**. Esta fuerza, puede medirse en la siguientes escalas no negativas: 
* $0< r \leq 0.3$ **Correlaci칩n d칠bil o nula**
* $ 0.3 < r < 0.7$ **Correlaci칩n moderada**  
* $0.7 \leq r \leq 1$ **Correlaci칩n fuerte**


El siguiente c칩digo nos permite hallar el coeficiente de correlaci칩n entre $X$ y $Y$

### 2.7 La prueba de significancia de la pendiente

Antes de hacer la prueba de la significancia, debemos definir lo que significa una relaci칩n lineal entre dos variables. 

**Definici칩n**
Se dice que existe una regresi칩n lineal entre dos variables continuas $X$ y $Y$ si existen constantes $\beta_{1}$ y $\beta_{0}$ tales que: 

```{math}
:class: my-math
  Y= \beta_{1}X+ \beta_{0}  

```

Por lo tanto, el objetivo de nuestro estudio  de la relaci칩n lineal entre $X$ y $Y$ se reducir칤a al de  probar las siguientes hip칩tesis de una cola:

```{math}
:class: my-math
  H_{0}:  \beta_{1} =0  \\ 
  H_{1}: \beta_{1} > 0 
```

donde 
* $H_{0}:\beta_{1}=0$ significa que **no existe relaci칩n lineal** entre $X$ y $Y$ 
* $H_{1}:\beta_{1}>0$ significa **existe una relaci칩n lineal positiva** entre $X$ e $Y$.

Para la prueba estadistica, usamos el estadistico $ t=  \frac{\beta_{1}}{S_{\beta_{1}}}$, que sigue una distribuci칩n t-student con $n-2$ grados de libertad,  donde:

* $S_{\beta_{1}}= \frac{s}{\sqrt{S_{xx}}}$

* $s = \sqrt{\frac{\sum_{i=1}^{n}(y_{i}-\widehat{y_{i}})}{n-2}}$

* $SS_{xx} = n\sum_{i=1}^{n}x_{i}^{2} - (\sum_{i=1}^{n}x_{i})^{2}$

Ya que la muestra es suficientemente grande, $n=1228 > 30$, la distribuci칩n del estadistico $t$ puede aproximarse a una distribuci칩n normal. Por lo tanto, para $n>30$, tenemos que $t_{\alpha , n-2}$ $\thickapprox  z_{\alpha}$.





### Nivel de confianza 
Para un nivel de confianza del 95%, elegimos $\alpha = 0.05$

### Criterios de rechazo
La hip칩tesis nula, $H_{0}$ se rechaza si el valor $p < 0.05 $ y el estadistico $t$ es positivo o tenemos $t > z_{0.05}$. 

### Recta de mejor ajuste
De la libreria *scipy* buscamos un modelo de regresi칩n lineal simple y ajustamos el modelo a los datos.

````python
import numpy as np
from scipy.stats import t

# 
x1 = new_df["Llamadas entrantes"].values  # Variable independiente
y1 = new_df["Tiempo de espera promedio (Segundos)"].values  # Variable dependiente
n = len(x1)  # Tama침o de la muestra (1228)

# 1. Ajustar el modelo de regresi칩n lineal
beta1, beta0 = np.polyfit(x1, y1, deg=1)  # beta1 = pendiente, beta0 = intercepto
````

### Errores
Calculamos los diferentes errores $s$, $SSE$, $SS_{xx}$ y $S_{\beta_{1}}$.

````python
# 2. Calculamos SSxx (Suma de Cuadrados de X)
x_mean = np.mean(x1)
SSxx = np.sum((x1 - x_mean) ** 2)  # 춰Notaci칩n corregida!

# 3. Calculamos residuos y Suma de Cuadrados de los Errores (SSE)
y_hat = beta0 + beta1 * x1
residuos = y1- y_hat
SSE = np.sum(residuos ** 2)  # Suma de cuadrados de los residuos

# 4. Error est치ndar de los residuos (s)
s_squared = SSE / (n - 2)  # Varianza residual
s = np.sqrt(s_squared)  # Error est치ndar de estimaci칩n

# 5. Error est치ndar de la pendiente (SE_beta1)
SE_beta1 = s / np.sqrt(SSxx)  # Usando SSxx
````

### Estad칤stico
Hallamos el estadistico $t$ de una sola cola e imprimimos los resultados.

````python
# 6. Estad칤stico t y p-valor
t_estadistico = beta1 / SE_beta1
p_valor_unilateral = 1 - t.cdf(t_estadistico, df=n-2)  # Solo cola derecha

# Resultados
print(f"Pendiente (1): {beta1:.4f}")
print(f"SSxx: {SSxx:.4f}")  # 
print(f"Error est치ndar de los residuos (s): {s:.4f}")
print(f"Error est치ndar de la pendiente (SE_1): {SE_beta1:.4f}")
print(f"Estad칤stico t: {t_estadistico:.4f}")
print(f"p-valor: {p_valor_unilateral:.4f}")
````

### Resultados
De los c치lculos anteriores tenemos que: 
* El estadistico $t=26.07 > z_{0.05} = 1.641$
* El valor $p < 0.05$  


Por lo tanto, rechazamos la hip칩tesis nula y concluimos que si hay significancia en la pendiente $\beta_{1}$.

### 2.8 Interpretaci칩n de los resultados
Del an치lisis anterior, observe que la pendiente $\beta_{1}$ es estadisticamente **significativa** (0.70), por lo que, s칤 existe una relaci칩n lineal positiva entre las variables. Sin embargo, observe que el coeficiente de $R^{2}$  es de 0.36, lo que significa que nada m치s el 36% de la variaci칩n lineal de $Y$ est치 explicada por la relaci칩n lineal entre $X$ y $Y$. Es decir, solo el 36% puede ser explicado por dicha relaci칩n, y el otro 64% se debe a la influencia de otros factores o variables no considerados. 

